<h3><a><center>Seaborn</center></a></h3>

Seaborn is also a visualization library built on top of Matplotlib. The library handles Pandas Dataframe very efficiently. Sea(Data) + Born(Infrences), getting infrences and knowledge from sea of data aka Big Data.

```python
# It's suggested to import pyplot module from matplotlib first. As Seaborn is built on top of matplotlib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# Data Import
data = pd.read_excel("Superstore Sales.xlsx")
df = data.sample(100).reset_index()
df.head()
```

##### Line Plot 

```python
# Seaborn takes data in the form of dataframe
# Seaborn gives values to x and y as defined in the arguments
sns.lineplot(x = "Order Date", y = "Sales", data = df)
plt.show()
```

```python
# We can plot the sales based on different categories using hue (column name for color encoding)
sns.lineplot(x = "Order Date", y = "Sales", data = df, hue = "Category" )
plt.show()
```

##### Histogram /Distribution plots 

```python
sns.distplot(df["Quantity"])
plt.show()
# With distplot we got histogram + line chart
```

```python
# library for statistical methods
from scipy.stats import norm
```

```python
# here in black line we can see normalized chart for shipping cost used in various problems.
sns.distplot(df["Shipping Cost"],hist= False,fit = norm)
plt.show()
```

##### Bar Plot 

```python
# the chart shows the category bsed on sales and in that it has filtered based on segments
sns.barplot(x = df.Category, y = df.Sales , hue = df.Segment)
plt.show()
```

##### Scatter Plot

```python
# Scatter plot is generally used for continuous variables. In the given chart we can see relaion between quantity and profit.
sns.scatterplot(x = "Quantity", y = "Profit", data = data , hue = "Category")
plt.show()
```

##### Heatmap - Representation of 2D- Data in Graphical Form where the values are encoded with respect to colors and it has specific meaning. 

```python
# 1 Heatmap is created for numerical values.
# 2 It is generaly used to find the correlation between different variables - Feature Engineering.
# 3 The following chart and their characteristics are defined based on the color bar.
sns.heatmap(df.corr())
```

```python
# We can change the minimum and maximum value and show the correlation value
sns.heatmap(df.corr(), vmin= 0, vmax= 1 , annot = True)
plt.show()
```

##### PairPlot - To plot pairwise relationship in dataset. The plots are generally bivariate in nature.

```python
sns.pairplot(df)
```
